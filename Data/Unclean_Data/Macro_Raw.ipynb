```python
import pandas as pd
import numpy as np
```

```python
df_macro = pd.read_excel("macro_raw_data.xlsx",sheet_name="macrodata")
df_finan = pd.read_excel("macro_raw_data.xlsx",sheet_name="financialdata")
```

```python
df_finan['date'] = pd.to_datetime(df_finan["year"].astype("str")+df_finan["month"].astype("str"),format="%Y%m")
df_finan.insert(0,'Date',df_finan["date"])
df_finan.drop(["year","month",'date'],axis=1,inplace=True)
df_macro['date'] = pd.to_datetime(df_macro["year"].astype("str")+df_macro["month"].astype("str"),format="%Y%m")
df_macro.insert(0,'Date',df_macro["date"])
df_macro.drop(["year","month",'date'],axis=1,inplace=True)
```

```python
df_macro.to_csv("../Macro_Raw.csv",index=None)
df_finan.to_csv("../Finan_Raw.csv",index=None)
```
